```python
import numpy as np
import copy
import os 
import re



file_name = "./model_results/32g_test/test_n_tasks_4_20__utilization_1_1__iter_40.log"


# Open the log file
with open(file_name, 'r') as file:
    lines = file.readlines()

# Initialize a dictionary to store configurations for each n_task group
configurations_res = {}

# Iterate through each line in the log file
for line in lines:

    # Split the line by "/"
    parts = line.split('/')
    # Extract the number of tasks (n_task)
    n_task = parts[1].split('_')[4]
    if n_task not in configurations_res.keys() : 
        configurations_res[n_task] = []
    # Extract the configurations
    config = parts[2]
    split_config = config[1:-1].split(',')
    # Append the configuration to the corresponding n_task group
    configurations_res[n_task].append(split_config)

print(configurations_res)
```
